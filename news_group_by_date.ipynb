```python
import pandas as pd
```

```python
start=1
end=856

for page in range(start,end+1):
    df=pd.read_csv(f"news_list_labeled_with_date/page{page}.csv")

    for idx,group in df.groupby("date"):
        try:
            pd.read_csv(f"news_list_group_by_date/{idx}.csv")
        except:
            f=open(f"news_list_group_by_date/{idx}.csv", "w")
            f.write("news,label,date\n")
            f.close()

        target_df=pd.read_csv(f"news_list_group_by_date/{idx}.csv")
        result=pd.concat([target_df,group])
        result.to_csv(f"news_list_group_by_date/{idx}.csv",index=False)
    
    print(f"page {page} finished!")
```

```python
import os
import datetime

files=os.listdir("news_list_group_by_date")

target_date_obj = datetime.datetime.strptime("2020.01.01", '%Y.%m.%d')

for file in files:
    date_str=file.split(".csv")[0]
    date_obj = datetime.datetime.strptime(date_str, '%Y.%m.%d')
    if date_obj<target_date_obj:
        os.remove(f"news_list_group_by_date/{file}")
```

```python
import datetime as dt

start_date_obj = dt.datetime.strptime("2020.01.01", "%Y.%m.%d").date()
end_date_obj = dt.datetime.strptime("2025.09.20", "%Y.%m.%d").date()

cur = start_date_obj
one_day = dt.timedelta(days=1)

no_news_dates=[]

while cur < end_date_obj:
    date_format=cur.isoformat().replace("-",".")
    try:
        pd.read_csv(f"news_list_group_by_date/{date_format}.csv")
    except:
        no_news_dates.append(date_format)
    cur += one_day
```

```python
for date in no_news_dates:
    f=open(f"news_list_group_by_date/{date}.csv", "w")
    f.write("news,label,date\n")
    f.close()
```
