```python
import pandas as pd
import os
```

```python
filenames=sorted(os.listdir("news_list_group_by_date"))
```

```python
filenames
```

```python
nums=0
max_nums=-1
min_nums=float('inf')
for filename in filenames:
    df=pd.read_csv(f"news_list_group_by_date/{filename}")
    temp=len(df["news"])
    max_nums=max(max_nums,temp)
    min_nums=min(min_nums,temp)
    nums+=temp
nums
```

```python
print(f"average : {nums/len(filenames)}")
print(f"max : {max_nums}")
print(f"min : {min_nums}")
```

```python
dates=[]
average_values=[]
for filename in filenames:
    df=pd.read_csv(f"news_list_group_by_date/{filename}")
    date=filename.split(".csv")[0]
    dates.append(date)
    avg_value=0
    try:
        avg_value=round(float(sum(df["label"].values)/len(df)),2)
    except:
        avg_value=0
    average_values.append(avg_value)
```

```python
df = pd.DataFrame({"dates": dates, "values": average_values})
df.to_csv(f"sentiment_average_value.csv", index=False)
```
